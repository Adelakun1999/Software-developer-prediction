```python
def shorten_categories(categories, cutoff):
    categorical_map ={}
    for i in range(len(categories)):
        if categories.values[i] >= cutoff:
            categorical_map[categories.index[i]] = categories.index[i]
        else:
            categorical_map[categories.index[i]] = 'Other'
    return categorical_map
```

```python
Country_map = shorten_categories(df['Country'].value_counts(), 400)
```

```python
df['Country'] = df['Country'].map(Country_map)
df['Country'].value_counts()
```

```python
def clean_education(x):
    if  'Bachelor’s degree' in x:
        return "Bachelor's degree"
    if 'Master’s degree' in x:
        return "Master's degree"
    if "Professional degree" in x or 'other doctoral' in x:
        return 'Post grad'
    return 'Less than a Bachelors'
```

```python
X = df.drop('Salary', axis=1)
y = df['Salary'].values
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV, RandomizedSearchCV
from sklearn.preprocessing import StandardScaler
```

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.tree import DecisionTreeRegressor, plot_tree
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, AdaBoostRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
X_train , X_test,y_train,y_test = train_test_split(X,y, test_size=0.3, random_state=101)
```

```python
scaler = StandardScaler()
X_train_scaler = scaler.fit_transform(X_train)
X_test_scaler = scaler.transform(X_test)
```

```python
import numpy as np
def model_building(model,X_train,X_test,y_train,y_test):
    model.fit(X_train,y_train)
    y_pred = model.predict(X_test)
    rmse = np.sqrt(mean_squared_error(y_test,y_pred))
    return f"The root mean squared error of {model} is {rmse:.2f}"
```

```python
linear_regression = LinearRegression()
```

```python
model_building(linear_regression, X_train_scaler,X_test_scaler,y_train,y_test)
```

```python
linear_regression.coef_
```

```python
lasso = Lasso()
model_building(lasso, X_train_scaler,X_test_scaler,y_train,y_test)
```

```python
ridge = Ridge()
model_building(ridge, X_train_scaler,X_test_scaler,y_train,y_test)
```

```python
tree = DecisionTreeRegressor(max_depth=3)
model_building(tree, X_train_scaler,X_test_scaler,y_train,y_test)
```

```python
plt.figure(figsize=(15,6),dpi=200)
plot_tree(tree,filled=True,feature_names=X.columns);
```

```python
rf_model =RandomForestRegressor()
model_building(rf_model, X_train_scaler,X_test_scaler,y_train,y_test)
```

```python
rf_model.feature_importances_
```

```python
pd.DataFrame(rf_model.feature_importances_, index=X.columns, columns=['feature importance']).plot(kind='barh', figsize=(10,5))
```

```python
param_grid = {'n_estimators':[int(x) for x in np.linspace(200,2000,10)],
             'max_features' : ['auto','sqrt','log2'],
             'max_depth': [int(x) for x in np.linspace(10,1000,10)],
             'min_samples_split':[2,5,10,14],
             'min_samples_leaf':[1,2,3,4,5,6,8]}

import warnings
warnings.filterwarnings('ignore')
```

```python
rf_random = RandomizedSearchCV(estimator=rf_model, param_distributions=param_grid,n_iter=20,n_jobs=-1,
                              scoring='neg_mean_squared_error',cv=5)
```

```python
model_building(rf_random, X_train_scaler,X_test_scaler,y_train,y_test)
```

```python
rf_random.best_params_
```

```python
regressor = rf_random.best_estimator_
```
